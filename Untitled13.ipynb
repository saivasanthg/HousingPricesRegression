```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
```

```python
data=pd.read_csv("trainhousing.csv")
data.head(100)
```

```python
y=data['SalePrice']
features=['LandContour','LotConfig','OverallQual','Neighborhood','OverallCond','YearBuilt','YrSold','RoofStyle','Exterior1st','Exterior2nd','MasVnrType','ExterQual','Foundation','BsmtQual','BsmtExposure','BsmtFinType1','BsmtFinSF1','BsmtUnfSF','HeatingQC','1stFlrSF','2ndFlrSF','GrLivArea','BsmtFullBath','BsmtHalfBath','FullBath','HalfBath','BedroomAbvGr','KitchenAbvGr','KitchenQual','TotRmsAbvGrd','Fireplaces','GarageCars','GarageArea','WoodDeckSF','OpenPorchSF','EnclosedPorch','3SsnPorch','ScreenPorch','SaleCondition']
```

```python
X=pd.get_dummies(data[features])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
print(X_train)
model=LinearRegression()
model.fit(X_train,y_train)
predict_=model.predict(X_test)
print(predict_)
```

```python
r2=r2_score(y_test,predict_)
output=pd.DataFrame({'SalePrice':predict_})
output
```
