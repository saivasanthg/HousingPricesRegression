```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.impute import SimpleImputer

```

```python
data=pd.read_csv('/content/AmesHousing.csv')
data.head()
```

```python
d=pd.DataFrame(data)
```

```python
column_names = d.columns.tolist()
print(column_names)
```

```python
corr_matrix=d.corr()
corr_matrix
```

```python
threshold = 0.7 # Adjust as needed
low_correlation_columns = corr_matrix['SalePrice'][abs(corr_matrix['SalePrice']) < threshold].index.tolist()
print("Columns below correlation threshold:")
for column in low_correlation_columns:
    print(column)
```

```python
df_filtered = d.drop(columns=low_correlation_columns)

print("\nFiltered DataFrame:")
print(df_filtered)
```

```python
categorical_columns = df_filtered.select_dtypes(include=['object']).columns
df_encoded = pd.get_dummies(df_filtered, columns=categorical_columns, drop_first=True)
df_cleaned = df_encoded.dropna()
```

```python
df_cleaned.isnull().any().any()
null_values = df_cleaned.isnull()
print(df_cleaned[null_values])
```
